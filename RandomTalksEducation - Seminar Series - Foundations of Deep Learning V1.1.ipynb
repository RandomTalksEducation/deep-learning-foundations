# Bir dersin sƒ±navana 12 saat √ßalƒ±≈üƒ±rsam ka√ß puan alƒ±rƒ±m?

Diyelim ki bir dersin sƒ±navƒ±na ge√ßmi≈üte giren √∂ƒürencilerin "ortalama √ßalƒ±≈üma saatleri" ve aldƒ±klarƒ± "Sƒ±nav Notu" a≈üaƒüƒ±daki tabloda verildiƒüi gibi olsun.

````python
import matplotlib.pyplot as plt
import numpy as np

# Veriler
X = np.array([2, 3, 4, 5, 6, 7, 8, 9, 10])
Y = np.array([50, 55, 60, 65, 70, 73, 77, 80, 82])

# Grafik olu≈üturma
plt.figure(figsize=(8, 5)) # Grafik boyutunu ayarlama

# Sa√ßƒ±lƒ±m grafiƒüi √ßizme
plt.scatter(X, Y, color='#007ACC', s=80, label='Sƒ±nav Puanlarƒ±')

# Ba≈ülƒ±k ve eksen etiketleri
plt.title('√áalƒ±≈üma Saati ve Sƒ±nav Puanƒ± ƒ∞li≈ükisi', fontsize=14, fontweight='bold')
plt.xlabel('√áalƒ±≈üma Saati (saat)', fontsize=12)
plt.ylabel('Sƒ±nav Puanƒ±', fontsize=12)

# X eksenini 1 saat aralƒ±klarla g√∂sterme
plt.xticks(np.arange(0, 12, 1))

# Y eksenini 40 ila 100 aralƒ±ƒüƒ±nda g√∂sterme
plt.yticks(np.arange(40, 101, 10))

# Izgara ekleme
plt.grid(True, linestyle='--', alpha=0.6)

# A√ßƒ±klama (legend) ekleme
plt.legend()

# Grafiƒüi g√∂sterme
plt.show()
````
